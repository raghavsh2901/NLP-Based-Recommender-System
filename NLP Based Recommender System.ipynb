```python
rates=pd.DataFrame(df1.groupby('title')['rate'].mean())
rates
```

```python
rates['numrate']=pd.DataFrame(df1.groupby('title')['rate'].count())
rates
```

```python
rates.sort_values(by='numrate',ascending=False)
```

```python
moviemat=df1.pivot_table(index='uid',columns='title',values='rate')
moviemat.head()
```

```python
swur=moviemat['Star Wars (1977)']
swur
```

```python
corr=moviemat.corrwith(swur)
corr
```

```python
corr_sw=pd.DataFrame(corr,columns=['Correlation Value'])
corr_sw.dropna(inplace=True)
corr_sw
```

```python
corr_sw.sort_values('Correlation Value',ascending=False)
corr_sw
```

```python
corr_sw1=corr_sw.join(rates['numrate'])
corr_sw1
```

```python
corr_sw1[corr_sw1['numrate']>100].sort_values(by='Correlation Value',ascending=False)
```

```python
ax=sns.heatmap(corr_sw1,vmin=-1,vmax=-1,center=0,cmap=sns.diverging_palette(20,220,n=200),square=True)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45,horizontalalignment='right');
```
